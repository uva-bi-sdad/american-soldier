```python
s106 = s106.fillna("")
```

```python
# define necessary functions
def pad(list_, padding, n_padding):
    n_padding = n_padding - len(list_)
    
    if n_padding <= 0:
        return list_
    
    for i in range(n_padding):
        list_.append(padding)
        
    return list_

def get_n_ids(subject_ids):
    return len(subject_ids.replace(" ", "").split(","))

def to_reformat(subject_ids):
    n_ids = get_n_ids(subject_ids)
    if n_ids > 2:
        return True
    return False

def reformat(data, row_id, new_data):
    # slice row from data
    row = data.loc[row_id, :]
    n_ids = get_n_ids(row['subject_id'])
    
    # regular pages
    subject_ids = row['subject_id'].replace(" ", "").split(",")[:2]
    subject_ids = ', '.join(subject_ids)
    
    image_names = row['image_name'].replace(" ", "").split(",")[:2]
    image_names = ', '.join(image_names)
    
    document_types = row['document_type'].split(",")[:5]
    document_types = ','.join(document_types)
    
    S106OH_series = row['S106OH.series'].replace(" ", "").split(",")[:2]
    S106OH_series = ', '.join(S106OH_series)
    
    S106OH_V54 = row['S106OH.V54'].replace(" ", "").split(",")[:2]
    S106OH_V54 = ', '.join(S106OH_V54)
    
    new_row = {'old_row_id': row_id, \
               'subject_id': subject_ids, \
               'image_name': image_names, \
               'document_type': document_types, \
               'S106OH.series': S106OH_series, \
               'S106OH.Q46.M': row['S106OH.Q46.M'], \
               'S106OH.Q46.M_marginalia': row['S106OH.Q46.M_marginalia'], \
               'S106OH.Q47.M': row['S106OH.Q47.M'], \
               'S106OH.Q47.M_marginalia': row['S106OH.Q47.M_marginalia'], \
               'S106OH.Q48A.M': row['S106OH.Q48A.M'], \
               'S106OH.Q48A.M_marginalia': row['S106OH.Q48A.M_marginalia'], \
               'S106OH.Q48B.F': row['S106OH.Q48B.F'], \
               'S106OH.Q49.F': row['S106OH.Q49.F'], \
               'S106OH.Q49.C_1': row['S106OH.Q49.C_1'], \
               'S106OH.Q49.C_2': "", \
               'S106OH.Q49.C_3': "", \
               'S106OH.Q49.C_4': "", \
               'S106OH.Q49.C_5': "", \
               'S106OH.V54': S106OH_V54}
    
    new_data = new_data.append(new_row, ignore_index = True)

    # contutation pages
    n_cont_pages = n_ids - 2

    subject_ids = pad(row['subject_id'].replace(" ", "").split(",")[2:], None, n_cont_pages)
    image_names = pad(row['image_name'].replace(" ", "").split(",")[2:], None, n_cont_pages)
    document_types = pad(row['document_type'].split(",")[5:], None, n_cont_pages)
    S106OH_series_vals = pad(row['S106OH.series'].replace(" ", "").split(",")[2:], None, n_cont_pages)
    S106OH_V54_vals = pad(row['S106OH.V54'].replace(" ", "").split(",")[2:], None, n_cont_pages)
    
    # iterate over continuation pages
    for i in range(n_cont_pages):
        # subject_ids
        subject_id = subject_ids[i]
        
        # image_name
        image_name = image_names[i]
        
        # document_type
        document_type = document_types[i]
        
        # S106OH.series
        S106OH_series = S106OH_series_vals[i]
        
        # S106OH.V54
        S106OH_V54 = S106OH_V54_vals[i]
        
        new_row = {'old_row_id': row_id, \
                   'subject_id': subject_id, \
                   'image_name': image_name, \
                   'document_type': document_type, \
                   'S106OH.series': S106OH_series, \
                   'S106OH.Q46.M': "", \
                   'S106OH.Q46.M_marginalia': "", \
                   'S106OH.Q47.M': "", \
                   'S106OH.Q47.M_marginalia': "", \
                   'S106OH.Q48A.M': "", \
                   'S106OH.Q48A.M_marginalia': "", \
                   'S106OH.Q48B.F': "", \
                   'S106OH.Q49.F': "", \
                   'S106OH.Q49.C_1': "", \
                   'S106OH.Q49.C_2': "", \
                   'S106OH.Q49.C_3': "", \
                   'S106OH.Q49.C_4': "", \
                   'S106OH.Q49.C_5': "", \
                   'S106OH.V54': S106OH_V54}


        if i == 0: # S106OH.Q49.C_2
            new_row['S106OH.Q49.C_2'] = row['S106OH.Q49.C_2']

        elif i == 1: # S106OH.Q49.C_3
            new_row['S106OH.Q49.C_3'] = row['S106OH.Q49.C_3']
            
        elif i == 2: # S106OH.Q49.C_4
            new_row['S106OH.Q49.C_4'] = row['S106OH.Q49.C_4']
            
        else: # S106OH.Q49.C_5
            new_row['S106OH.Q49.C_5'] = row['S106OH.Q49.C_5']
            
        new_data = new_data.append(new_row, ignore_index = True)
    return new_data
```

```python
reformat_ids = [to_reformat(x) for x in s106['subject_id']]
new_data = pd.DataFrame(columns=['old_row_id', 'subject_id', 'image_name', 'document_type', 'S106OH.series', 'S106OH.Q46.M', 'S106OH.Q46.M_marginalia', 'S106OH.Q47.M', 'S106OH.Q47.M_marginalia', 'S106OH.Q48A.M', 'S106OH.Q48A.M_marginalia', 'S106OH.Q48B.F', 'S106OH.Q49.F', 'S106OH.Q49.C_1', 'S106OH.Q49.C_2', 'S106OH.Q49.C_3', 'S106OH.Q49.C_4', 'S106OH.Q49.C_5', 'S106OH.V54'])
for i in range(len(reformat_ids)):
    if reformat_ids[i]:
        new_data = reformat(s106, i, new_data)
```
