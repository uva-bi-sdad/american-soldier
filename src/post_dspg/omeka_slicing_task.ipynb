```python
def slice(path, file):
    df = pd.read_csv(path + file + ".csv", index_col = 0)
    print(file, len(df))
    i = 0
    k = 0
    
    while i < len(df) - 250:
        j = i + 250
        print("{i}:{j}".format(i = i, j = j))
        
        # slice data frame
        tmp = df[i:j]
        
        # save sliced data
        tmp.to_csv("../../data/post_dspg/to_import_copy_sliced/{file}_{k}.csv".format(file = file, k = k))
        
        # increase variables
        i += 250
        k += 1
        
    if (i < len(df)):
        print("{i}:".format(i = i))
        tmp = df[i:]
        tmp.to_csv("../../data/post_dspg/to_import_copy_sliced/{file}_{k}.csv".format(file = file, k = k))
```

```python
def get_file(full_path):
    file = full_path.split("/")[-1]
    file = file.split(".")[0]
    return file

def get_path(full_path):
    path_to_file = full_path.split("/")[:-1]
    path_to_file = "/".join(path_to_file)
    path_to_file += "/"
    return path_to_file
```
